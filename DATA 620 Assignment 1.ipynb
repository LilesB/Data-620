```python
# Import the networkx and matplotlib modules
# Implemented 

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline 

'''As a fan of the HBO series, The Wire, I decided to design a graph which mirrors the Barksdale hierarchy'''
# Create the Barskdale graph
b_graph = nx.Graph()

# Add edges to the graph object
# Using add_edges_from tuples were created 
b_graph.add_edges_from([
    ("Avon","Stringer"),
    ("Avon","Levy"),
    ("Avon","Brianna"),
    ("Avon","Wee-Bey"),
    ("Stringer","Shamrock"),
    ("Shamrock","Slim Charles"),
    ("Slim Charles","Country"),
    ("Slim Charles","Tank"),
    ("Slim Charles","Perry"),
    ("Slim Charles","Rico"),
    ("Slim Charles","Tater"),
    ("Slim Charles","Cutty"),
    ("Slim Charles","Gerard"),
    ("Slim Charles","Sapper"),
    ("Slim Charles","Bernard"),
    ("Slim Charles","Bodie"),
    ("Bodie","Poot"),
    ("Bodie","Sterling"),
    ("Bodie","Puddin"),
    ("Poot","Sterling"),
    ("Poot","Puddin")
])

#Draw the resulting graph
nx.draw(b_graph, with_labels=True, font_weight="bold", node_size = 600, node_color = "skyblue", alpha = 0.75)
plt.draw()
plt.show()
```
