## 3 Taking the most common words, how many unique words represent half of the total words in the corpus?

```python
halfWrds = nltk.FreqDist(text9) # used to encode "frequency distributions"
topWrds = halfWrds.most_common(200)
```

```python
i = 0
wrds = []
for w in topWrds:
    if i >= len(text9):
        break
    else:
        i = i + w[1]
        wrds.append(w)
len(wrds)
```

```python
# Create a dataframe which will host the words and the number of times it is present
wrdsDF = pd.DataFrame(wrds, columns = ['word','number'])
wrdsDF
```
